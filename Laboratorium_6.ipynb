### Zadanie 1

Za pomocą biblioteki pandas odczytaj z pliku inflacja.xlsx:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel('inflacja.xlsx', sheet_name='Arkusz1')
df
```

```python
df.info()
```

1. Najwyższy i najniższy wskaźnik inflacji


```python
max = df.Wartość.max()
min = df.Wartość.min()
```

```python
print("Najwyższy wskaźnik inflacji :",max)
```

```python
print("Najniższy wskaźnik inflacji :",min)
```

2. Rok w którym był najwyższy i najniższy wskaźnik inflacji

```python
year_max = int(df["Rok"][df["Wartość"] == df["Wartość"].max()].values)
year_max

print(f"Najwyższy wskaźnik inflcji : {max} był w roku: {year_max}")
```

```python
year_min = int(df["Rok"][df["Wartość"] == df["Wartość"].min()].values)
year_min

print(f"Najwyższy wskaźnik inflcji : {min} był w roku: {year_min}")
```

3. Lata, dla których inflacja była większa od średnij inflacji

```python
mean = df.Wartość.mean()
print("Średnia inflacji : ", mean)
```

```python
above_mean = df[df['Wartość'] > mean].reset_index(drop=True)
above_mean
```

```python
years_above_mean = list(above_mean.Rok)
years_above_mean
```

```python
x_values = df["Rok"]
y_values = df["Wartość"]
to_legend = "Score of inflation"

# plt.figure(figsize = (15,8))
plt.plot(x_values, y_values, marker = "o", color="blue", label = to_legend, linewidth=1,)
plt.xlabel("Year")
plt.ylabel("Score of inflation")
plt.title("Score of Inflation during the years from data file inflacja.xlsx")
plt.legend(loc = "best")
plt.show()
```

1. W razie potrzeby usuń wielowymiarowe nazwy kolumn( przykład spłaszczania MultiIndexów [https://www.pauldesalvo.com/how-to-flatten-multiindex-columns-into-a-single-index-dataframe-in-pandas/](https://www.pauldesalvo.com/how-to-flatten-multiindex-columns-into-a-single-index-dataframe-in-pandas/))

```python
df = df.replace(",", ".", regex = True)
df = df.replace(' ','', regex = True)
df = df.replace('b', np.nan, regex = True)
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.info()
```

2. Sprawdź, czy dane w kolumnach Minimalne wynagrodzenie ustawowe i Przeciętne wynagrodzenie zostały wczytane prawidłowo i zmień typ kolumn na float64.

```python
pd.to_numeric(df["Minimalne wynagrodzenie ustawowe"], downcast='float')
pd.to_numeric(df["Przeciętne wynagrodzenie"], downcast='float')
df
```

3. Oblicz stosunek Przeciętne wynagrodzenie / Minimalne wynagrodzenie ustawowe i otrzymany wynik pokaż na wykresie.

```python
stosunek = df["Przeciętne wynagrodzenie"]/df["Minimalne wynagrodzenie ustawowe"]
stosunek
```

```python
type(stosunek)
```

```python
plt.figure(figsize = (12,6))
plt.xticks(df["Rok"])
plt.ylabel("Przeciętne wynagrodzenie / Minimalne wynagrodzenie ustawowe")
plt.xlabel("Rok")

plt.plot(df["Rok"], stosunek, marker="o")
```
